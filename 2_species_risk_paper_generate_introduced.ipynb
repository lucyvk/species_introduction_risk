```python
import csv
out_location = "paper_webs"
```

```python
######## FIXED PARAMETER SET FOR INTRODUCED SPECIES ###################################

# Leads to 45 options for intermediate species (in_1) (5 ni's x 3 ris x 3 cis)
# Leads to 45 options for top predator species (in_2) (5 ni's x 3 ris x 3 cis)
# Leads to 729 options for two species together (3 nis x 3 nis x 3 ris x 3 ris x 3 cis x 3 cis)

in_1_nis = [0.4,0.45,0.5,0.55,0.6] # intermediate niche parameter values
in_2_nis = [0.8,0.85,0.9,0.95,1] # high niche parameter values
xs = [0.3,0.5,0.7] # for generating small, medium, and large feeding range
#xs = [0.5,0.75,1] # more extreme version of introduced species feeding range (for SI figures)
cs = [0,1,2] # for generating low, middle, high center of feeding range 

in_uid = 333

# Files to save the introduced species parameters 
in_1_file = open(f"./{out_location}/introduced_1.csv",'w')
#in_1_file = open(f"./{out_location}/introduced_1_extreme.csv",'w')
in_1_writer = csv.writer(in_1_file,dialect="excel")
in_2_file = open(f"./{out_location}/introduced_2.csv",'w')
#in_2_file = open(f"./{out_location}/introduced_2_extreme.csv",'w')
in_2_writer = csv.writer(in_2_file,dialect="excel")
in_both_file = open(f"./{out_location}/introduced_both.csv",'w')
#in_both_file = open(f"./{out_location}/introduced_both_extreme.csv",'w')
in_both_writer = csv.writer(in_both_file,dialect="excel")

# Write header rows
in_1_writer.writerow(['in_id','ni','ri','ci'])
in_2_writer.writerow(['in_id','ni','ri','ci'])
in_both_writer.writerow(['in_id1','in_id2','ni1','ni2','ri1','ri2','ci1','ci2'])

in_1_temp = []
in_2_temp = []

for i in range(0,len(in_1_nis)):
    
    for j in range(0,len(xs)):
        
        for k in range(0,len(cs)):

            #generate invader parameters and write to file
            in_1_ni = in_1_nis[i]
            in_2_ni = in_2_nis[i]

            x = xs[j]
            in_1_ri = round(x*in_1_ni,5)
            in_2_ri = round(x*in_2_ni,5)

            c_flag = cs[k]
            # start of range
            if c_flag == 0:
                in_1_ci = round(in_1_ri/2,5)
                in_2_ci = round(in_2_ri/2,5)
            # middle of range
            elif c_flag == 1:
                in_1_ci = round(((in_1_ni-(in_1_ri/2))/2+in_1_ri/2),5)
                in_2_ci = round(((in_2_ni-(in_2_ri/2))/2+in_2_ri/2),5)
            # top of range
            else:
                in_1_ci = in_1_ni
                in_2_ci = in_2_ni

            #give each potential invader a unique id that is different
            in_1_id = in_uid
            in_uid += 1
            in_2_id = in_uid
            in_uid += 1

            in_1_writer.writerow([in_1_id,in_1_ni,in_1_ri,in_1_ci])
            in_2_writer.writerow([in_2_id,in_2_ni,in_2_ri,in_2_ci])

            if i%2 == 0:
                in_1_temp.append([in_1_id,in_1_ni,in_1_ri,in_1_ci])
                in_2_temp.append([in_2_id,in_2_ni,in_2_ri,in_2_ci])        

for one in in_1_temp:
    for two in in_2_temp:
        in_both_writer.writerow([one[0],two[0],one[1],two[1],one[2],two[2],one[3],two[3]])

in_1_file.close()
in_2_file.close()
in_both_file.close()  
```
