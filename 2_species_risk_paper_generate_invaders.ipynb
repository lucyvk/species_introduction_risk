```python
import csv
out_location = "paper_webs_0624"
dummy_location = "NONE"
```

```python
######## FIXED PARAMETER SET FOR INVADERS ###################################

# Leads to 45 options for invader 1 (5 ni's x 3 ris x 3 cis)
# Leads to 45 options for invader 2 (5 ni's x 3 ris x 3 cis)
# Leads to 729 options for invader 1 and 2 (3 nis x 3 nis x 3 ris x 3 ris x 3 cis x 3 cis)

inv_1_nis = [0.4,0.45,0.5,0.55,0.6]
inv_2_nis = [0.8,0.85,0.9,0.95,1]
# xs = [0.5,0.75,1] # more extreme version of introduced species parameters
xs = [0.3,0.5,0.7]
cs = [0,1,2]

inv_uid = 333
b_flag = 2
    
#names of files changed below so not overwritten
inv_1_file = open('./' + out_location + '/invader_1.csv','w')
inv_1_writer = csv.writer(inv_1_file,dialect="excel")
inv_2_file = open('./' + out_location + '/invader_2.csv','w')
inv_2_writer = csv.writer(inv_2_file,dialect="excel")
inv_both_file = open('./' + out_location + '/invader_both.csv','w')
inv_both_writer = csv.writer(inv_both_file,dialect="excel")

#write header rows
inv_1_writer.writerow(['inv_id','ni','ri','ci'])
inv_2_writer.writerow(['inv_id','ni','ri','ci'])
inv_both_writer.writerow(['inv_id1','inv_id2','ni1','ni2','ri1','ri2','ci1','ci2'])

inv_1_temp = []
inv_2_temp = []

for i in range(0,len(inv_1_nis)):
    
    for j in range(0,len(xs)):
        
        for k in range(0,len(cs)):

            #generate invader parameters and write to file
            inv_1_ni = inv_1_nis[i]
            inv_2_ni = inv_2_nis[i]

            x = xs[j]
            inv_1_ri = round(x*inv_1_ni,5)

            x = xs[j]
            inv_2_ri = round(x*inv_2_ni,5)

            c_flag = cs[k]
            # start of range
            if c_flag == 0:
                inv_1_ci = round(inv_1_ri/2,5)
                inv_2_ci = round(inv_2_ri/2,5)
            # middle of range
            elif c_flag == 1:
                inv_1_ci = round(((inv_1_ni-(inv_1_ri/2))/2+inv_1_ri/2),5)
                inv_2_ci = round(((inv_2_ni-(inv_2_ri/2))/2+inv_2_ri/2),5)
            # top of range
            else:
                inv_1_ci = inv_1_ni
                inv_2_ci = inv_2_ni

            #give each potential invader a unique id that is different
            #from the ids of all the existing species
            inv_1_id = inv_uid
            inv_uid += 1
            inv_2_id = inv_uid
            inv_uid += 1

            inv_1_writer.writerow([inv_1_id,inv_1_ni,inv_1_ri,inv_1_ci])
            inv_2_writer.writerow([inv_2_id,inv_2_ni,inv_2_ri,inv_2_ci])

            if i%b_flag == 0:
                inv_1_temp.append([inv_1_id,inv_1_ni,inv_1_ri,inv_1_ci])
                inv_2_temp.append([inv_2_id,inv_2_ni,inv_2_ri,inv_2_ci])        

for one in inv_1_temp:
    for two in inv_2_temp:
        inv_both_writer.writerow([one[0],two[0],one[1],two[1],one[2],two[2],one[3],two[3]])

inv_1_file.close()
inv_2_file.close()
inv_both_file.close()  
```
